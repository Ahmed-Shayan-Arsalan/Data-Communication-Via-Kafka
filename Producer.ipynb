```python
producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                         value_serializer=lambda x:
                         json.dumps(x).encode('utf-8'))
```

```python
import yfinance as yf
stocks = yf.download('SI=F')
stocks
```

```python
for i in range(0,stocks.shape[0]):
    stocks1 = stocks.iloc[[i]].to_dict(orient="records")[0]
    stocks2 = dict((':'.join(k),v) for k,v in stocks1.items())
    producer.send('cyan1',value=stocks2)
```

```python
for i in range(0, stocks.shape[0]):
    stocks1 = stocks.iloc[[i]].to_dict(orient="records")[0]
    stocks2 = {':'.join(k): float(v) if isinstance(v, str) and v.replace('.', '', 1).isdigit() else v for k, v in stocks1.items()}
    producer.send('cyan3', value=stocks2)
```
