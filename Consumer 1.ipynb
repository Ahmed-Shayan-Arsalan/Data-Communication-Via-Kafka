```python
from kafka import KafkaConsumer
from json import dumps,loads
import json
```

```python
consumer = KafkaConsumer('cyan1',
                        bootstrap_servers=['localhost:9092'],
                        value_deserializer=lambda x:
                        json.loads(x.decode('utf-8')))
```

```python
for message in consumer:
    print(message.value)
    # print(message.value)
```
